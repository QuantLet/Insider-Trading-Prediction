```python
import joblib

file_path = "/root/autodl-tmp/DATA/data_bundles.pkl"

data_bundles = joblib.load(file_path)

print("读取完成")
print("包含的 set:", data_bundles.keys())
```

```python
import pandas as pd

def combine_train_test(bundle):
    return pd.concat(
        [bundle['train'], bundle['test']],
        axis=0,
        ignore_index=True
    )

set1_full = combine_train_test(data_bundles['set1'])
set2_full = combine_train_test(data_bundles['set2'])
set3_full = combine_train_test(data_bundles['set3'])
```

```python
check1 = (
    set1_full[['Stkcd','Trddt']].values ==
    set2_full[['Stkcd','Trddt']].values
).all()

check2 = (
    set1_full[['Stkcd','Trddt']].values ==
    set3_full[['Stkcd','Trddt']].values
).all()
```

```python
df1 = set1_full[['Stkcd','Trddt','insider_trading']].copy()
df2 = set2_full[['Stkcd','Trddt','insider_trading']].copy()
df3 = set3_full[['Stkcd','Trddt','insider_trading']].copy()

df1 = df1.rename(columns={'insider_trading':'Y1'})
df2 = df2.rename(columns={'insider_trading':'Y2'})
df3 = df3.rename(columns={'insider_trading':'Y3'})
```

```python
df_joint = df1.merge(df2, on=['Stkcd','Trddt'], how='inner')
df_joint = df_joint.merge(df3, on=['Stkcd','Trddt'], how='inner')
```

```python
joint_table = (
    df_joint
    .groupby(['Y1','Y2','Y3'])
    .size()
    .reset_index(name='count')
    .sort_values(['Y1','Y2','Y3'])
)

print(joint_table)
```
